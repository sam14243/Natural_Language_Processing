```python
import nltk
from nltk.util import ngrams
from nltk.metrics.distance import edit_distance
import pandas as pd
import numpy as np
# !pip install textdistance
import textdistance
import re
from collections import Counter
```

```python
def find_best_fit_word(sentence, choices):
    n = len(sentence.split())
    sentence_ngrams = list(ngrams(sentence.split(), n))
    best_fit_word = None
    min_distance = float('inf') 
    for choice in choices:
        choice_ngrams = list(ngrams(choice.split(), n))
        distance = edit_distance(sentence_ngrams, choice_ngrams) 
        if distance < min_distance:
            min_distance = distance
            best_fit_word = choice 
    return best_fit_word
```

```python
find_best_fit_word("Mr Patrick is our new", ["principal", "principle"])
```

```python
find_best_fit_word("The company all the terms.", ["accepted", "excepted"])
```
