```python
import os

TEXT_PATH = "./SemEval2017/docsutf8/"
KEY_PATH = "./SemEval2017/keys/"
```

```python
import pandas as pd
import os

file_list = os.listdir(TEXT_PATH)
df = pd.DataFrame(columns=['File', 'Content'])

for file in file_list:
    file_path = os.path.join(TEXT_PATH, file)
    with open(file_path, 'r') as f:
        content = f.read()

    df.loc[len(df)] = {'File': file, 'Content': content}
df
```

```python
file_list = os.listdir(KEY_PATH)
nested_list = []

for file_name in df['File']:
    file_path = os.path.join(KEY_PATH, file_name[:-3] + 'key')
    
    with open(file_path, 'r') as f:
        lines = [line.strip() for line in f.readlines()]
        
        nested_list.append(lines)
nested_list
```

```python
df["Keys"] = nested_list
df
```

```python
import spacy
import pytextrank
# example text

def extract_keywords(text):
    nlp = spacy.load("en_core_web_sm")
    nlp.add_pipe("textrank")
    doc = nlp(text)
    return [phrase.text for phrase in doc._.phrases[:10]]
```

```python
pred = []
for i in range(len(df)):
    pred += [extract_keywords(df["Content"][i])]

df["Predicted"] = pred
```
